```python
def visualize(image, w=7, h=7):
    plt.figure(figsize=(w, h))
    plt.axis('off')
    plt.imshow(image)
```

```python
# Load images
image1 = cv2.imread('tree.jpg')
image2 = cv2.imread('cats_dogs.jpg')
```

```python
# from BGR to RGB
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# resize to (256, 256)
image1 = cv2.resize(image1, (256, 256))
image2 = cv2.resize(image2, (256, 256))
```

```python
# visualize them
visualize(image1, 5, 5)
visualize(image2, 5, 5)
```

```python
random.seed(7)

aug = A.HorizontalFlip(p=0.5)(image=image1)['image']
visualize(aug, 4, 4)
```

```python
aug = A.CLAHE(clip_limit=15, always_apply=True)(image=image1)['image']
visualize(aug, 4, 4)
```
