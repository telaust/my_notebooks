```python
# from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
df = pd.read_csv('data.csv').drop('id', axis=1).drop('Unnamed: 32', axis=1)
```

```python
xtrain = df.iloc[:400, 1:]
xtest = df.iloc[400:, 1:]

ytrain = df.iloc[:400, 0]
ytest = df.iloc[400:, 0]
```

```python
df
```

```python
tree = DecisionTreeClassifier('entropy').fit(xtrain, ytrain)
```

```python
tree.predict(xtest)
```

```python
predicted = tree.predict(xtest)
```

```python
plot = plot_tree(tree, filled=True)
```

```python
def get_accuracy(predicted, real):
    concurrence = 0
    for i in range(real.size):
        if real.iloc[i] == predicted[i]:
            concurrence += 1
    return concurrence / real.size
```

```python
print(f'accuracy = {get_accuracy(predicted, ytest):.3f}')
```
