```python
import torch
from torch import nn
from torchvision.datasets import MNIST
from torchvision import transforms
from torch.utils.data import DataLoader
from torch import optim
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# Define hyperparams
n_epoches = 30
n_batches = 128
lr = 1e-2
```

```python
img_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,) ) # мат ожидание и дисперсия 
])
```

```python
dataset = MNIST('./data', transform=img_transform, download=True)
```

```python
loader = DataLoader(dataset=dataset, batch_size=n_batches, shuffle=True)
```

```python
# 784 - 300 - 150 - 300 - 784
class AE(nn.Module):
    def __init__(self, inp_size, hid1_size, hid2_size):
        super(AE, self).__init__()
        
#         self.inp_size = inp_size
#         self.hid1_size = hid1_size
#         self.hid2_size = hid2_size
        
        self.lin1 = nn.Linear(inp_size, hid1_size)
        nn.init.kaiming_uniform_(self.lin1.weight) # He init
        self.hid1 = nn.ELU() 
        
        self.lin2 = nn.Linear(hid1_size, hid2_size)
        nn.init.kaiming_uniform_(self.lin2.weight) # He init
        self.hid2 = nn.ELU() 
        
        self.lin3 = nn.Linear(hid2_size, inp_size)
        
    def forward(self, x):
        x = self.hid1(self.lin1(x)) #   alternative:      x = nn.ELU(self.lin1(x))
        x = self.hid2(self.lin2(x)) #   alternative:      x = nn.ELU(self.lin2(x))
        x = self.lin3(x)
        
        return x

    
# simple autoencoder
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        
        self.encoder = nn.Sequential(nn.Linear(28*28, 128),
                                    nn.ReLU(True),
                                    nn.Linear(128, 64),
                                    nn.ReLU(True),
                                    nn.Linear(64, 12),
                                    nn.ReLU(True),
                                    nn.Linear(12, 3))
        
        
        self.decoder = nn.Sequential(nn.Linear(3, 12),
                                    nn.ReLU(True),
                                    nn.Linear(12, 64),
                                    nn.ReLU(True),
                                    nn.Linear(64, 128),
                                    nn.ReLU(True),
                                    nn.Linear(128, 28*28),
                                    nn.Tanh())
    
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        
        return x
```

```python
ae = AE(784, 300, 150)
```

```python
# define criterion & optimizer
crit = nn.MSELoss() 
opt = optim.Adam(ae.parameters(), lr=lr, weight_decay=1e-5)
```

```python
losses = []

# TODO: обернуть этот цикл в отдельную функцию

for ep in range(n_epoches):
    start_ep = time.time() # measure how long is passing one epoch
    for data in loader:
        img, _ = data
        
        img = img.view(img.size(0), -1)
        
        
        # forward pass
        out = ae.forward(img)
        loss = crit(out, img)
        
        
        # backward pass
        opt.zero_grad()
        loss.backward()
        opt.step()
        
    end_ep = time.time() - start_ep 
    
    losses.append(loss.item())
    if ep % 10 == 0 and ep != 0:
        print("loss is {} at epoch = {}; time for that epoch is {.4f} sec".format(loss.item(), ep, end_ep))
            
```

```python
# save it
torch.save(ae.state_dict(), 'simple_autoencoder.pth')
```
