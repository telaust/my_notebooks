```python
import requests
import pprint
from bs4 import BeautifulSoup
import re
```

```python
regex = re.compile('[^a-zA-Z$]')
longwords = re.compile(r'\W*\b\w{16,400}\b')
shortwords = re.compile(r'\W*\b\w{1,3}\b')
```

```python
# English words

en_url = "https://en.wikipedia.org/wiki/Earth"
en_page = requests.get(en_url)

soup = BeautifulSoup(en_page.content, 'html.parser')

without_spaces = soup.text.strip()

s = without_spaces.replace("[", "").replace("]", "").replace("\n", "") 

# https://stackoverflow.com/questions/3900054/python-strip-multiple-characters

str = ''.join([i for i in s if not i.isdigit()]).replace('"', '').replace('^', "")


# bad_chars = '().,\/'
clear = ''.join([i for i in str if (i.isalpha() or i.isspace())])


clear = regex.sub(' ', s)

text = longwords.sub('', clear).lower().replace("wikipedia", "").replace("wikimedia", "")


good_text = shortwords.sub('', text) # слова в нижнем регистре, длина каждого от 4 до 15 букв

en_words = good_text.split()

with open("data/english_words.txt", "w") as file:
    for i in range(len(en_words)):
        file.write(en_words[i] + '\n')
file.close()
```

```python
# French words

fr_url = "https://fr.wikipedia.org/wiki/France"
fr_page = requests.get(fr_url)

soup = BeautifulSoup(fr_page.content, 'html.parser')

fr_regex  = re.compile('[^a-zA-Z$]')
fr_clear = fr_regex.sub(' ', soup.text)

temp_str_fr = longwords.sub('', shortwords.sub('', fr_clear)).lower().replace("wikipedia", "").replace("wikimedia", "")
fr_words = temp_str_fr.split()


# write French vocabulary

with open("data/french_words.txt", "w") as file:
    for i in range(len(fr_words)):
        file.write(fr_words[i] + '\n')
file.close()
```
